```python
dementia = pd.read_csv("dataset/dementia_dataset.csv")
dementia.head()
```

# fix balance using SMOTE

```python
X = dementia.drop(columns=['Group','Subject ID', 'MRI ID'])
y = dementia['Group']
columns = X.columns


oenc = OrdinalEncoder()
imputer = SimpleImputer(strategy='median')
X = oenc.fit_transform(X)
X = imputer.fit_transform(X)
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
df = pd.DataFrame(X, columns=columns)
df['Group'] = y
```
